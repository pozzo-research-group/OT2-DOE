```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
full_data = pd.read_csv(r"C:\Users\Edwin\Downloads\04_09_21 Cadmium Oleate HT Blanked")
full_data
```

```python
wavelength_columns = [col for col in full_data.columns if 'nm' in col]
wavelengths = [float(wavelength.replace("nm","")) for wavelength in wavelength_columns]
spectras = full_data[wavelength_columns]
```

```python
for i, row in spectras.iterrows():
    row = list(row)
    plt.plot(wavelengths, row)
plt.xlim(350,450)
plt.ylim(-0.05, 1)
plt.legend()
plt.xlabel('Wavelength nm')
plt.ylabel('AU')
```

```python
selected_data = full_data[full_data['Time Sonicated']==0]
labels = selected_data['ZnOl 84mM volpct']
labels
labels = ['ZnOl '+ str(label)+' volume %'for label in labels]
```

```python
# print(len(labels))
selected_spectras = selected_data[wavelength_columns]
selected_spectras
for j, (i, row)in enumerate(selected_spectras.iterrows()):
    row = list(row)
    plt.plot(wavelengths, row, label = labels[j])
    
plt.xlim(350,500)
plt.ylim(-0.05, 1)
plt.legend()
plt.xlabel('Wavelength nm')
plt.ylabel('AU')
```
