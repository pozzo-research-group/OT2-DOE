```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
abs_data = pd.read_csv(r"C:\Users\Edwin\Desktop\05_03_21_CdTOP_Blanked_Abs")

# def function identifier
wavelength_columns = [col for col in abs_data.columns if 'nm' in col]
wavelengths = [float(wavelength.replace("nm","")) for wavelength in wavelength_columns]

abs_spectra = abs_data[wavelength_columns]
```

```python
labels = abs_data['Sonication Time']
for j, (i, row) in enumerate(abs_spectra.iterrows()):
    absorbances = list(row) 
    plt.plot(wavelengths, absorbances, label = 'Sonication Time = ' + str(labels[j]) + 'min')
    
plt.xlim(400,420)
plt.xlabel('Wavelength nm')
plt.ylabel('AU')
plt.legend()
# plt.ylim(-0.1, 0.1)
```

```python
fl_data = pd.read_csv(r"C:\Users\Edwin\Desktop\05_03_21_CdTOP_Blanked_FL")

# def function identifier
wavelength_columns = [col for col in fl_data .columns if 'nm' in col]
wavelengths = [float(wavelength.replace("nm","")) for wavelength in wavelength_columns]

fl_spectra = fl_data[wavelength_columns]

fl_spectra
```

```python
labels = fl_data['Sonication Time']
for j, (i, row) in enumerate(fl_spectra[0:4].iterrows()):
    fl = []
    for f in list(row):
        if type(f) == str:
            f = f.replace('*', '')
        else: 
            f = f 
        f = float(f)
        fl.append(f) 
    plt.plot(wavelengths, fl, label = 'Sonication Time = ' + str(labels[j]) + 'min')
    
#     print(fl)
# plt.xlim(370,600)
plt.xlabel('Wavelength nm')
plt.ylabel('RFU')
plt.legend()
# plt.ylim(0, 35000)
```
